# Gradient Descent Algorithm

```python
# Input parameters
start_x=16
learning_rate=0.1
n_iterations=100
# ----------------

# Store the minimum gradient results
minimum = []

# Set x to the starting point x (satrt_x)
x = start_x

for i in range(n_iterations):
    # Calculate gradient (derivative of xÂ²)
    gradient = 2 * x

    # Update x by moving opposite to the gradient
    x = x - learning_rate * gradient

    print(f"Iteration {i}: x = {x:.4f}, f(x) = {x*x:.4f}")
    minimum.append(x)
```

```python
import matplotlib.pyplot as plt

# Plots the training and validation loss curves
plt.title('Gradient Desccent Line Plot')
plt.xlabel('Iteration')
plt.ylabel('Gradient')
plt.plot(minimum, label='Minimum')
plt.show()
```
